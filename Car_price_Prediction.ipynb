```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
import csv
import warnings
warnings.filterwarnings("ignore")
```

```python
Car_prices = pd.read_csv("/content/drive/MyDrive/carprices.csv")
Car_prices
```

```python
# car models are text instead of numbers. Change to dummies
dummies = pd.get_dummies(Car_prices["Car Model"], drop_first=True)
dummies

```

```python
# Define features and target
X = pd.concat([Car_prices[["Mileage", "Age(yrs)"]], dummies], axis=1)
y = Car_prices["Sell Price($)"]
```

```python
# Train model
model=LinearRegression()
model.fit(X,y)
```

```python
# 1) Predict price of a mercedez benz that is 4 yr old with mileage 45000
model.predict([[45000,4,0,0]])
```

```python
# 2) Predict price of a BMW X5 that is 7 yr old with mileage 86000
model.predict([[86000,7,0,1]])
```

```python
# 3) Tell me the score (accuracy) of your model.
model.score(X,y)
```
